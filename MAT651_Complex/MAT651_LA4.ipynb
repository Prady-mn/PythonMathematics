```python
import numpy as np
import sympy as sp
from typing import *
import import_ipynb
from MAT651_LA3 import checkCR,checkHarmonic;
import matplotlib.pyplot as plt
```

```python
def mySympify(f:str,vars = {"x":sp.symbols("x",real = True),"y":sp.symbols("y",real = True)}) -> ["function","x","y"]:
    f = sp.sympify(f,locals = vars)
    x,y = list(vars.values())
    return [f,x,y]
```

```python
def checkConformal(f:str):
    if(not checkCR(f)[0]):
        return {"CR":False}
    f,x,y = mySympify(f)
    df = sp.diff(sp.re(f),x) + 1j*sp.diff(sp.im(f),x)
    critical = sp.solve(sp.Eq(df,0),x,y)
    #if(df == 0):
    return {"Conformal": (df != 0),"Critical Points":critical,"Function":f}
```

```python
checkConformal("cos(y)*e**x + 1j*sin(y)*e**x")
```

```python
checkConformal("x**2 + 2*x*y*1j - y**2")
```

```python
#asin(cosh(-y)*cos(x) + 1j*sinh(-y)*sin(x))
checkConformal("cosh(-y)*cos(x) + 1j*sinh(-y)*sin(x)")
```

## Reflection and Translation

> WAP to plot the reflection along Real and Imaginary axis

```python
def reflectX(z):
    return z.real - 1j*z.imag
def reflectY(z):
    return -z.real + 1j*z.imag

def plotReflections(z):
    fig = plt.figure();
    for i in z:
        r = np.array([i,reflectX(i),reflectY(i)])
        plt.scatter(r.real,r.imag, label = i)
    plt.grid()
    plt.legend()
    plt.axhline(color = "black")
    plt.axvline(color = "black")
    return plt
```

```python
reflectX(1+1j)
```

```python
plotReflections([1+1j, 2-3j])
```

> WAP to reflect a triangle

```python
def reflectShape(z):
    z.append(z[0])
    z = np.array(z)
    
    fig = plt.figure()
    
    plt.plot(z.real,z.imag,label = "Original")
    z = np.vectorize(reflectX)(np.vectorize(reflectY)(z))
    plt.plot(z.real,z.imag,label = "Reflected")
    
    plt.grid()
    plt.legend()
    plt.axhline(color = "black")
    plt.axvline(color = "black")
    return plt
```

```python
reflectShape([2+2j,-1+8j,-2+1j]);
```

## Linear and Bilinear Transformations

**TRANSLATION**

```python
def translateShape(z,z0):
    z.append(z[0])
    z = np.array(z)
    
    fig = plt.figure()
    
    plt.plot(z.real,z.imag,label = "Original")
    z += z0
    plt.plot(z.real,z.imag,label = "Translated")
    
    plt.grid()
    plt.legend()
    plt.axhline(color = "black")
    plt.axvline(color = "black")
    return plt
```

```python
translateShape([2+1j,-1+4j,-2+1j],-1-0.5j)
```

```python
def translateShapePolar(z,z0):
    z.append(z[0])
    z = np.array(z)
    
    fig = plt.figure()
    
    plt.polar(np.angle(z),np.abs(z),label = "Original")
    z += z0
    plt.polar(np.angle(z),np.abs(z),label = "Translated")
    
    plt.legend()
    return plt
```

```python
translateShapePolar([5+5j,1+1j,-2+3j],2+3j)
```

```python
translateShapePolar([np.exp(1j*np.pi/2),np.exp(1j*np.pi),np.exp(3j*np.pi/2),np.exp(2j*np.pi)],-1+1.5j)
```

**ROTATION**

```python
def rotation(z,theta):
    return np.abs(z)*np.exp(1j*(np.angle(z)+theta))
def rotateShapePolar(z,theta):
    z.append(z[0])
    z = np.array(z)
    
    fig = plt.figure()
    
    plt.polar(np.angle(z),np.abs(z),label = "Original")
    z = np.vectorize(rotation)(z,theta)
    plt.polar(np.angle(z),np.abs(z),label = "Rotated")
    
    plt.legend()
    return plt
```

```python
rotateShapePolar([-2-1j,-1-1j,1+1j],np.pi/2)
```

```python
def rotateShape(z,theta):
    z.append(z[0])
    z = np.array(z)
    
    fig = plt.figure()
    
    plt.plot(z.real,z.imag,label = "Original")
    z = np.vectorize(rotation)(z,theta)
    plt.plot(z.real,z.imag,label = "Rotated")
    
    plt.legend()
    plt.grid()
    plt.legend()
    plt.axhline(color = "black")
    plt.axvline(color = "black")
    return plt
```

```python
rotateShape([-2-1j,-1-1j,1+1j],np.pi/2)
```

**Bilinear Transformation**

```python
def bilinear(z,a,b,c,d):
    return (a*z+b)/(c*z+d)
    
def bilinearShape(z,a,b,c,d):
    if(b*c - a*d == 0):
        raise Exception("BILINEAR")
    z.append(z[0])
    z = np.array(z)
    
    fig = plt.figure()
    
    plt.plot(z.real,z.imag,label = "Original")
    z = np.vectorize(bilinear)(z,a,b,c,d)
    plt.plot(z.real,z.imag,label = f"Transformed{(a,b,c,d)}")
    
    plt.legend()
    plt.grid()
    plt.legend()
    plt.axhline(color = "black")
    plt.axvline(color = "black")
    return plt
```

```python
bilinearShape([-2-1j,-1-1j],2,1j,-1j,2)
```
