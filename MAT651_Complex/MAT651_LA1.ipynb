> 3D Ploting

```python
z = np.linspace(-20,20,500)
x = np.sin(z)
y = np.cos(z)

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection = "3d");
p1 = ax.scatter3D(x,y,z,cmap = "Dark2",c = (z))
fig.colorbar(p1,shrink = 0.8,label = "z value")
plt.show()
```

```python
x = np.linspace(-2,2,100);
y = lambda X,i: [[np.cos(x),
                  np.exp(x),
                  x**2,
                  np.tan(x)][i] for x in X]
functions = ["y = cos(x)","y = e^x","y = x^2","y = tan(x)"]

plt.figure(figsize = (12,8))
for i in range(0,4):
    plt.subplot(2,2,i+1)
    plt.grid()
    plt.plot(x,y(x,i),label = functions[i])
    plt.legend(fontsize = 12,loc = 1)
    plt.axhline(color = 'black')
    plt.axvline(color = 'black')

plt.show()
```
