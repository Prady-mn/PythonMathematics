> To plot various 2D and 3D graphs

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

**Sine and Cosine Graphs**

```python
x = np.linspace(-2*np.pi,2*np.pi,200)
y_sin = np.sin(x)
y_cos = np.cos(x)

fig = plt.figure(figsize = (14,8))
plt.plot(x,y_sin,label = "sin x")
plt.plot(x,y_cos,label = "cos x")
plt.axhline(color = "black")
plt.axvline(color = "black")
plt.xticks()
plt.grid()
plt.legend()
plt.show()
```

**Linear, Quadratic and Exponential**

```python
x = np.linspace(-5,5,200);
t = np.linspace(-np.pi,np.pi,200)
y_lin = 11*x + 8;
y_quad = 7*x**2 + 4*x + 1
#y_quad = x**2
y_exp = np.e**x
x_circ,y_circ = 4*np.cos(t),4*np.sin(t)

fig = plt.figure(figsize = (8,8))
plt.plot(x,y_lin,label = "Linear")
plt.plot(x,y_quad,label = "Quadratic")
plt.plot(x,y_exp,label = "Exponential")
plt.plot(x_circ,y_circ,label = "Circle")
plt.xlim(left = -10,right = 10)
plt.ylim(bottom = -10, top = 10)
plt.axhline(color = "black")
plt.axvline(color = "black")
plt.xticks()
plt.grid()
plt.legend()
plt.show()
```

**6 Pointed Star**

**Surface of Sphere**

```python
u = np.linspace(0,2*np.pi,100) #alpha
v = np.linspace(0,np.pi,100) # beta, gamma

r = 10

x = r*np.outer(np.cos(u),np.sin(v))
y = r*np.outer(np.sin(u),np.sin(v))
z = r*np.outer(np.ones(np.size(u)),np.cos(v))

fig = plt.figure(figsize = (14,10))
ax = plt.axes(projection = "3d");
ax.plot_surface(x,y,z,cmap = 'Greens')
plt.show()
```

**Sinusoidal Surface in 3D**

```python
xx,yy = np.meshgrid(range(-10,10),range(-10,10))
zz = np.sin(xx+yy)
fig = plt.figure(figsize = (14,10))
ax = plt.axes(projection = "3d");
ax.plot_surface(xx,yy,zz,cmap = "Purples")
plt.show()
```

**Helix Plot**

```python
z = np.linspace(-20,20,500)
x = np.sin(z)
y = np.cos(z)

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection = "3d");
p1 = ax.scatter3D(x,y,z,cmap = "Dark2",c = (x*y)**2)
plt.show()
```

```python
z = np.linspace(-20,20,500)
x = np.sin(z)
y = np.cos(z)

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection = "3d");
p1 = ax.scatter3D(x,y,z,cmap = "Dark2",c = (z))
fig.colorbar(p1,shrink = 0.8,label = "z value")
plt.show()
```

**Plane in 3D**

```python
xx,zz = np.meshgrid(range(-20,20),range(-15,15))
yy = np.ones(np.shape(xx))
fig = plt.figure(figsize = (14,10))
ax = plt.axes(projection = "3d");
p1 = ax.plot_surface(xx,yy,5*zz,cmap = "Spectral",label = "XY plane")
p2 = ax.scatter3D(zz,xx,yy,cmap = "winter_r",c = yy+xx,label = "XZ plane")
fig.colorbar(p1,shrink = 0.8,label = "XY Plane")
fig.colorbar(p2,shrink = 0.8,label = "XZ Plane")
plt.show()
```

**Deformed Parabola Surface in 3D**

```python
xx,yy = np.meshgrid(range(-20,20),range(-20,20))
zz = -xx**2 + yy**2
fig = plt.figure(figsize = (14,10))
ax = plt.axes(projection = "3d");
ax.plot_surface(xx,yy,5*zz,cmap = "cividis",label = "XY plane")
plt.show()
```

**3D Parabola**

```python
yy,zz = np.meshgrid(range(-10,10),range(-10,10))
xx = 4*(yy**2 + zz**2)
fig = plt.figure(figsize = (14,10))
ax = plt.axes(projection = "3d");
ax.plot_surface(xx,yy,zz,cmap = "turbo",label = "XY plane")
plt.show()
```

**3D Hyperbola**

```python
yy,xx = np.meshgrid(range(-10,10),range(-10,10))
zz = (1 - 2*(yy**2 - xx**2))**0.5
fig = plt.figure(figsize = (14,10))
ax = plt.axes(projection = "3d");
ax.plot_surface(xx,yy,zz,cmap = "ocean")
plt.show()
```

**Function on a Meshgrid: Contour Graph**

```python
def f1_z(x,y):
    return np.sin((x**2 + y**2))

def f2_z(x,y):
    return np.sin((1 - 2*(y**2 - x**2))**0.5)**2

fig = plt.figure(figsize = (10,10))
x,y = np.meshgrid(range(-5,5),range(-5,5))
plt.contour(x,y,f1_z(np.cos(x),y),cmap = 'winter')
plt.colorbar()
#plt.contour(x,y,f2_z(x,y),colors = 'red')
plt.show()
```

```python
def f2_z(x,y):
    return np.sin((1 - 2*(y**2 - x**2))**0.5)**2

fig = plt.figure(figsize = (10,10))
x,y = np.meshgrid(range(-5,5),range(-5,5))
plt.contour(x,y,f2_z(x,y),colors = 'red')
plt.colorbar()
plt.show()
```

```python
def f3_z(x,y):
    return np.sin(x)**10 + np.cos(10 + x*y)*np.sin(x)

fig = plt.figure(figsize = (10,10))
x,y = np.meshgrid(np.linspace(-5,5,20),np.linspace(-5,5,20))
plt.contourf(x,y,f3_z(x,y),10,cmap = 'Greens')
plt.colorbar()
plt.show()
```

```python
def f4_z(x,y):
    return (np.sin(x) + np.cos(y)**2)

fig = plt.figure(figsize = (10,10))
x,y = np.meshgrid(np.linspace(0,20,20),np.linspace(0,20,20))
contour1 = plt.contourf(x,y,f4_z(x,y),10,cmap = 'RdGy')
plt.colorbar()
plt.clabel(contour1)
plt.show()
```

> Plot of Heat Map

```python
def f5_z(x,y):
    return (x**2 + y**2)

fig = plt.figure(figsize = (10,10))
x,y = np.meshgrid(np.linspace(0,10,20),np.linspace(0,10,20))
contour1 = plt.contourf(x,y,f5_z(x,y),50,cmap = 'RdYlBu')
plt.colorbar()
plt.clabel(contour1)
plt.title("Heat Radiated from Portion of Sphere")
plt.show()
```

> Plot of Isotherm

```python
def f6_z(x,y):
    return np.e**(-(x/3)**2-(y/3)**2) + np.e**(-(x+2)**2-(y+2)**2)
fig = plt.figure(figsize = (10,10))
x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))
contour1 = plt.contour(x,y,f6_z(x,y),10,cmap = "Set2") 
plt.colorbar()
plt.clabel(contour1)
plt.show()
```
