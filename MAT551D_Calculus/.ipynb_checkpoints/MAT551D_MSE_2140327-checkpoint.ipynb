> WAP to explain meshgrid with syntax and example.

Meshgrid is a numpy function that takes two numpy arrays and creates a grid corresponding to both the arrays taking one along x-axis and the other along y-axis. <br>
$\quad$ Syntax: numpy.meshgrid(x = numpy.array,y = numpy.array)<br>
The numpy array contains all the points to be created. range can also be given as an input.

```python
X = range(-5,5);
Y = range(-2,2);

mesh = np.meshgrid(X,Y);
```

```python
plt.scatter(mesh[0],mesh[1])
plt.title("Plot of Mesh Grid")
plt.show()
```

> WAP to explore the behaviour of helix.<br>
$\quad \quad r(t) = (cos at)i+ (sin at)j+ (bt)k$<br>
$\quad t:~-10~\text{to}~10$

```python
#t = np.linspace(-10,10)
mesh = np.meshgrid(np.linspace(-10,10),np.linspace(-10,10))

def plotHelix(a,b):
    t = np.linspace(-10,10,100)
    rx = np.cos(a*t);
    ry = np.sin(a*t);
    rz = b*t;
    fig = plt.figure(figsize = (5,5));
    ax = plt.axes(projection = "3d");
    ax.plot3D(rx,ry,rz,);
```

```python
plotHelix(1,2)
```

```python
plotHelix(1,5)
```

```python
plotHelix(4,2)
```

Inference: <br>
'a' controls the coiling of the helix. When 'a' increases the number of turns in the helix for a same range increases. <br>
'b' controls the height expansion of the helix. When 'b' increases, the height of the helix increase, ie. helix is stretched along the z-axis more and more.

> WAP to plot given contour and level curves.

```python
def f1(x,y):
    return 100 - x**2 - y**2;
levelcurves = [0,51,75];
X,Y = np.meshgrid(np.linspace(-10,10),np.linspace(-10,10));
Z = 100 - X**2 - Y**2

fig = plt.figure(figsize = (7,7));
ax = plt.axes(projection = '3d')

ax.plot_surface(X,Y,Z)

for i in levelcurves:
    ax.plot_surface(X,Y,np.ones(np.shape(X))*i)
```
