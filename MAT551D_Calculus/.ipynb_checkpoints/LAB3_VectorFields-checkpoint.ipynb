**Space Curve**
> When a particle moves through sapce during a time interval $I$, we think of the particles coordinates as functions defined on $I$, as $\ldots$<br>
$\quad x = f(t) \quad y = g(t) \quad z = h(t) \quad \forall~t \in I = [a,b]$<br>
The points $(x,y,z)$ are the coordinates of the particle in the interval $I$.

```python
t = np.linspace(-10,10,400);
x = np.cos(t)
y = np.sin(t)
z = t

fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection = "3d");
ax.scatter3D(x,y,z,cmap = "ocean_r")
plt.show()
```

**Plotting Vector Fields**

> Quiver Plot with one Arrow

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
ax.axis([-3,3,-3,3])
ax.quiver(0,0,1,1)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
ax.quiver([0,0],[0,0],[2,1],[2,-1])
ax.axis([-2,2,-2,2])
```

**Slope Field**

```python
import matplotlib.pyplot as plt

xx,yy = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))
u,v = xx,yy

plt.quiver(xx,yy,u,v,angles = 'xy')
plt.axhline(color = 'black')
plt.axvline(color = 'black')
plt.show()
```
