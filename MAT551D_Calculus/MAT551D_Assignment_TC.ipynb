```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
def plotContour_3D(z,limit,clines):
    x,y = np.meshgrid(range(-limit,limit+1),range(-limit,limit+1))
    fig = plt.figure(figsize=(10,4))
    ax = fig.add_subplot(1,2,1)
    ax.contour(x,y,z(x,y),clines)

    ax = fig.add_subplot(1, 2, 2, projection='3d')
    ax.plot_surface(x,y,z(x,y))
```

```python
plotContour_3D(lambda x,y: np.cos(x)*np.cos(y)*np.exp(-(x**2 + y**2)**0.5/4),10,20)
print("a - 33")
```

```python
plotContour_3D(lambda x,y: -(x*y**2)/(x**2 + y**2),5,30)
print("b - 36")
```

```python
plotContour_3D(lambda x,y: 1/(4*x**2 + y**2),5,10)
print("c - 34")
```

```python
plotContour_3D(lambda x,y: np.exp(-y)*np.cos(x),5,10)
print("d - 35")
```

```python
plotContour_3D(lambda x,y: (x*y*(x**2 - y**2))/(x**2 + y**2),5,20)
print("e - 32")
```

```python
plotContour_3D(lambda x,y: -y**2 * y**4 * x**2,10,20)
print("f - 31")
```
