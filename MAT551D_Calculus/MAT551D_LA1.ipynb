> WAP to plot the following$\ldots$<br>
> - Scatter Plot of: P(1,2),Q(in second quadrant),R(in third),S(in fourth)

```python
import matplotlib.pyplot as plt

points = {'P':(1,2) , 'Q':(-3,1) ,'R':(-5,-4) ,'S':(2,-5)}
#X = [1,-3,-5,2]
#Y = [2,1,-4,-5]
for i,(x,y) in points.items():
    plt.scatter(x,y)
    plt.annotate(i+str(points[i]),(x,y))
plt.axhline(color = 'black')
plt.axvline(color = 'black')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.grid()
plt.show()
```

> 3D Plotting

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection = "3d");

points = [(2,3,-1),(-3,4,2),(2,-3,1)]

randpoints = []
for i in range(50):
    randpoints.append(((np.random.randint(-50,50)),(np.random.randint(-50,50)),(np.random.randint(-50,50))))
    
ax.plot3D(np.array(randpoints).T[0],np.array(randpoints).T[1],np.array(randpoints).T[2])
plt.show()
```

> WAP to plot a 3D helix

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

ax = plt.axes(projection = "3d");

z = np.linspace(-20,20,500)
x = np.sin(z)
y = np.cos(z)

ax.scatter3D(x,y,z,cmap = "Dark2",c = (x*y)**2)
plt.show()
```
