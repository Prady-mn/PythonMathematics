**Find the Equation of Osculating Circle**

> WAP to find the unit tangent vector, curvature and principle unit normal vector of $y = x^2$. Plot the curve **$y(x)$**, 
the **curvature function**, and the **osculating circle** at point **$(x,y)$**

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

#X,Y = eval(input("Enter Point:"))
X0,Y0 = 0,0

i,j,t,r, = symbols('i,j,t,r,')
direction = {i:1,j:1}

x,y = t,t**2
r = x*i + y*j

dr = collect(diff(r,t),(i,j))
mod_dr = ((dr.coeff(i)**2 + dr.coeff(j)**2)**0.5)
T = simplify(dr/mod_dr)
#print(T)

k = (diff(T,t)/mod_dr).subs({t:0}).subs(direction)
radius = 1/k

theta = symbols('theta')

#(radius*cos(t), radius*sin(t)+radius), (t, -pi, pi)

plotting.plot(r.subs(direction),T.subs(direction))
plotting.plot_parametric((radius*cos(t), radius*sin(t)+radius), (t, -pi, pi),xlim = (-10,10))
```
