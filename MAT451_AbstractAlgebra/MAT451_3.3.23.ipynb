> Given the demand equation for a monopolist$\left(p\right)$ and the cost function, WAP to find the marginal profit, 
where $\ldots$<br>
$\quad\quad p\left(x\right) = 100 - 0.01x \quad C\left(x\right) = 50x + 10000$
>> Marginal Profit is found as follows-<br>
    Differentiate Profit function: $P^{'}\left(x\right)$ <br>
    $\quad P\left(x\right) = xp -  C\left(x\right)$

```python
from sympy import *

x = Symbol("x")
C_fn = 50*x + 10000
p_fn = 100 - 0.01*x

P_fn = (p_fn*x - C_fn).simplify()

mP_fn = diff(P_fn)

print("Cost Function :",C_fn)
print("Demand Function :",p_fn)
print("Profit Function :",P_fn)
print("Marginal Profit Function :",mP_fn)
```

> Given the cost function$\left(C\right)$, WAP to find marginal cost and marginal profit at $x = 50$.<br>
$\quad\quad p\left(x\right) = 20 \quad C\left(x\right) = 100 + 8x + 0.1x^{2}$

```python
from sympy import *

x = Symbol("x")
p_fn = 20
C_fn = 100 + 8*x + 0.1*(x**2)

mC_fn = diff(C_fn)
P_fn = x*p_fn - C_fn
mP_fn = diff(P_fn)

print("At x = 20:")
print("Marginal Cost = ",int(mC_fn.subs({x:50})))
print("Marginal Profit = ",int(mP_fn.subs({x:50})))
```

#### LA6: Differential Equations Using Sympy

> WAP to solve the following DE-<br>
$\quad \dfrac{d}{dx}y\left(x\right) = 1$

```python
from sympy import *

#Independent Variable
x = Symbol("x")

#Dependent Variable
y = Function("y")(x)

#Assigning Differential Equation
D = Eq(y.diff(x),1)

#Printing The Equation
print(D)
display(D)

#Solving the Equation
dsolve(D,y)
```

> WAP to solve the following DE-<br>
$\quad \dfrac{dp}{dt} - r = 0$

```python
from sympy import *

t = Symbol("t")
p = Function("p")(t)
r = Symbol("r")
D = Eq(p.diff(t,1)-r,0)

print(D)
display(D)
dsolve(D)
```

#### Initial value problems

> $dy/dx  = -0.3y , y(0) = 5$

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(y,x):
    dydx = -0.3*y
    return dydx

y0 = 5
x = arange(0,20,0.5)

y1 = odeint(model,y0,x)

plt.plot(x,y1)
plt.show()
```

> $dy/dx  = 2k , y(0) = 1$

```python
def model(y,t,k):
    dydx = 2*k
    return dydx
y0 = 1
t = linspace(0,20,50)

k=1
yk1 = odeint(model,y0,t,args=(k,))
k=2
yk2 = odeint(model,y0,t,args=(k,))
k=5
yk3 = odeint(model,y0,t,args=(k,))

plt.plot(t,yk1,'r')
plt.plot(t,yk2,'b.-')
plt.plot(t,yk3,'k+')
plt.show()
```


> $dy/dx  = 1-y , y(0) = 0$

```python
from numpy import *
from scipy.integrate import odeint
import matplotlib.pyplot as plt

def model(y,x):
    dydx = 1-y
    return dydx

y0 = 0
x = arange(0,20,0.5)

y1 = odeint(model,y0,x)

plt.plot(x,y1)
plt.show()
```

Solve the differential equation $y^n + 2y' + y = 0$ , y(0)=5, y'(0) = -3

```python
def du_dx(u,x):
    return [u[1],-2*u[1]-u[0]]

u0 = [5,-3]
xs = linspace(0,10,200)
us = odeint(du_dx,u0,xs)
ys = us[:,0]
plt.plot(xs,ys,color = "black")
plt.title("Solution of $y'' + 2y' + y = 0$")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### SLOPE FIELDS
> The differential equation says that the slope of a solution curve at a point (x,y) on the curve is F(x,y). If we draw short line segments with slope F(x,y) at several points, the result is called a direction field (or slope field)

Plot the slope field for dy/dx = x+y

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3,3,0.3)
y = np.arange(-3,3,0.3)
X,Y = np.meshgrid(x,y)
dy = X+Y
dx = np.ones(dy.shape)
color = dy
plot1 = plt.figure()
plt.quiver(X,Y,dx,dy, color = 'green')
plt.title("Slope field for $dy/sx = x+y$")
plt.show(plot1)
```
