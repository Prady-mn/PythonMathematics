> WAP to define the Composites of two functions- <br>
$\qquad f(x) = e^{x} + x^{2}$ <br>
$\qquad g(x) = 2x + 1$

```python
from sympy import *
from math import e

def f(x):
    return e**x + x**2;

def g(x):
    return 2*x + 1;

def compositeSum(f,g):
    return lambda x: f(x) + g(x);

def compositeMinus(f,g):
    return lambda x: f(x) - g(x);

def compositeMult(f,g):
    return lambda x: f(x) * g(x);

def compositeDiv(f,g):
    return lambda x: f(x) / g(x);

def compositeComp(f,g):
    return lambda x: f(g(x));

funcSum = compositeSum(f,g);
funcMinus = compositeMinus(f,g);
funcMult = compositeMult(f,g);
funcDiv =  compositeDiv(f,g);
funcComp = compositeComp(f,g);

x = Symbol("x");
print("The Functions Are- ")
print("f(x) = ",f(x));
print("g(x) = ",g(x));
print("(f+g)(x) = ",funcSum(x));
print("(f-g)(x) = ",funcMinus(x));
print("(f*g)(x) = ",funcMult(x));
print("(f/g)(x) = ",funcDiv(x));
print("f(g(x)) = ",funcComp(x));

print();

print("Evaluated at x = 1")
print("f(2) = ",f(1));
print("g(2) = ",g(1));
print("(f+g)(2) = ",funcSum(1));
print("(f-g)(2) = ",funcMinus(1));
print("(f*g)(2) = ",funcMult(1));
print("(f/g)(2) = ",funcDiv(1));
print("f(g(2)) = ",funcComp(1));
```

> WAP determine *LHL* and *RHL* -
> 1. $f\left(x\right) = x^{2} + 2x + 1 , x = 5$
> 2. $f\left(x\right) = 2x + 4 , x = 4$

```python
import numpy as np;

def getLimit(f):
    n = int(input("Enter value of Limit Point: "));
    
    L = np.linspace(n-0.1,n,20);
    R = np.linspace(n+0.1,n,20);
    
    lhl = f(L);
    rhl = f(R);
    
    print("From Left Hand Side- ");
    print(lhl)
    print("From Right Hand Side- ");
    print(rhl)
    print("At the Point - ");
    print(f(n))
    
f1 = lambda x: x**2 + 2*x + 1
print(f1)
getLimit(f1);
print("--------------------------------------------------------------------------");
f2 = lambda x: 2*x + 4
print(f2)
getLimit(f2);
```

Limit Piece-wise Functions

> WAP to define a piece-wise function and find it's LHL and RHL. Print Limit if it exists.<br>
$
\quad \quad
\begin{equation}
    f\left( x \right) = \left\{
    \begin{array}{cc}
        x^{2} + 1 \quad x \geq 1 \\
           x - 1 \quad x < 1
    \end{array}
    \right\}
\end{equation}
$

```python
import numpy as np;

f = lambda x: x**2 + 1 if(x >= 1) else x - 1

n = 1;

L = np.linspace(n-0.1,n,20);
R = np.linspace(n+0.1,n,20);

lhl = [f(i) for i in L];
rhl = [f(i) for i in R];
print()
print("--------------------------------------------------------------------------");
print("From Left Hand Side- ");
print(lhl);
print("--------------------------------------------------------------------------");
print("From Right Hand Side- ");
print(rhl)
print("--------------------------------------------------------------------------");
if(lhl[-1] == rhl[-1]):
    print("Since LHL = RHL, limit exists");
    print("f(",n,") = ",f(n));
else:
    print("Limit Does Not Exist");
    print("LHL = ",lhl[-1],"    RHL = ",rhl[-1]);
```

Maxima and Minima of Functions<br>
> Find first derivative<br>
> Find critical points as- $\dfrac{dy}{dx} = 0$<br>
> Evaluate $\dfrac{d^{2}y}{dx^{2}}$ at all critical point<br>
> Check sign of second drivative:<br>
> $\quad \quad $Positive: Minima $\quad$ Negative: Maxima

> WAP to find maxima and minima of single variable function: <br>
> $\quad $

```python
from sympy import *
import numpy as np

def getCritical(f):
    f_1 = diff(f,x);
    X = solve(f_1,x)
    return X;

#Returns the following
# Maxima: True
# Minima: False
def checkMaxMin(f,p):
    f_2 = diff(f,x,2);
    if(f_2.subs({x:p}) > 0):
        return False
    return True

def findMaxMin(f):
    X = getCritical(f);
    maxima,minima = [],[];
    for i in X:
        #print(i,type(i))
        if(i.is_real != True):
            continue;
        if(checkMaxMin(f,i)):
            maxima.append(i)
        else:
            minima.append(i)
    return maxima,minima

def printMaxMin(f):
    maxima,minima = findMaxMin(f)
    print("Maxima: ",maxima);
    print("Minima: ",minima);

x = Symbol("x")
f = x**4 
printMaxMin(f);
```

### Applications of Derivatives in Economics
$\quad$ It is knowns that if $ y = f\left(x\right)$, then the derivaive can be interpreted as the rate of change of $y$ with respect to $x$. Here, we examine the applications of this idea in Economics.

#### Cost Function
$\quad$ Consider a company which produced a certain commodity. Then, $C\left(x\right)$ is the total const of the commodity in producing $x$ units. It can be represented as follows$\dots$<br>
$\quad\quad C\left(x\right) = a + bx + cx^{2} + dx^{3}$ <br>
where $\ldots$<br>
$\quad\quad\bullet$ $a$ - overhead cost like rent, maintanance $($commodities independent of $x~)$ <br>
$\quad\quad\bullet$ $b$ - cost of raw materials $($commodities propotional of $x~)$<br>
$\quad\quad\bullet$ $c,d$ - labour, overtime, inefficiencies, etc. $($represented by $x~)$<br>

#### Marginal Cost Function
$\quad$ If the number of items produced is incresed from $x_{1}$ to $x_{2}$, then the additional cost is$\ldots$<br>
$\quad\quad C \left( x_{1} \right) - C\left(x_{2}\right)$<br>
Then the average rate of change of cost is$\dots$<br>
$\quad \quad \dfrac{\Delta C}{\Delta x} = \dfrac{C \left(x_{1} + \Delta x\right) - C\left(x_{1}\right)}{\Delta x}$<br>
The limit of this quantity as $\Delta x \rightarrow 0$, ie. the instantaneous rate of change of cost is$\dots$<br>
$\quad \quad \lim_{x\to 0} \dfrac{\Delta C}{\Delta x} = \dfrac{dC}{dx}$<br>
Taking $\Delta x = 1$ and very large $n$,(since $x$ in integer), we get\dots <br>
$\quad \quad C^{'}\left(n\right) \approx C\left(n + 1\right) - C\left(n\right)$<br>
Thus, the marginal cost of producing $n$ units is approximately equal to produing the next unit.

#### Revenue Function
$\quad$ We define $R\left(x\right)$ as the revenue generated by selling $x$ units. The derivative, $R^{'}\left(x\right)$ is the marginal revenue. This is used to measure the rate of increase in revenue per unit increase in sales. If $x$ units are sold at a price $p$ per unit, then the total revenue is given as$\ldots$<br>
$\quad \quad R\left(x\right) = x \cdot p$

#### Profit Function
$\quad$ We can define a profit function by knowing cost and revenue function, $P \left(x\right)$. It is given as$\ldots$ <br>
$\quad \quad P\left(x\right) = R\left(x\right) - C\left(x\right)$

> The differential equation says that the slope of a solution curve at a point $\left( x,y\right)$ on the curve is $F\left(x,y\right)$. If we draw short line segments with slope $F\left(x,y\right)$ at several points, the result is called a direction field (or slope field)

> WAP to plot the slope field of $\dfrac{dy}{dx} = x + y$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-3,3,0.3)
y = np.arange(-3,3,0.3)
X,Y = np.meshgrid(x,y)
dy = X+Y
dx = np.ones(dy.shape)
color = dy
plot1 = plt.figure()
plt.quiver(X,Y,dx,dy, color = 'green')
plt.title("Slope field for $\dfrac{dy}{dx} = x+y$")
plt.show(plot1)
```

**SIMPLE AND COMPOUND INTEREST**
> $\text{SI} = \dfrac{P \cdot T \cdot R}{100} $<br>
> $\text{CI} = P \cdot \left(\left(1 + \dfrac{r}{100}\right)^{t} - 1\right) $<br>
> $ \text{A} = P + I$<br>

```python
def simpleInt(P,T,R):
    return P*T*R/100

def compoundInt(P,T,R):
    return P*((1 + R/100)**T - 1)

def getValue():
    return [int(input("Enter Principal:")),int(input("Enter Time:")),int(input("Enter Rate:"))]

PTR = getValue()
SI = simpleInt(*PTR)
print("Simple Interest:",SI)
CI = compoundInt(*PTR)
print("Compound Interest:",CI)
```

> In a bank, principle increases continously at the rate of $5%$ per year. An amount of Rs.$1000$ is deposited with the bank. WAP to find the worth after 10 years. Also find time to double the initial amount.

```python
from sympy import *
import numpy as np

rate = 0.1
principle = 1000
time = 5
p,t,r,A = symbols("p,t,r,A")
CI = Eq(A,p*(1 + r)**t)

myData = {p:principle, r:rate, t:time, A:A}
display(CI.subs(myData))

myData[t] = t
myData[A] = 2*principle
display(Eq(t,solve(CI.subs(myData))[0]))
```

> EXPONENTIAL GROWTH

> If a population of a town increases from 50,000 to 1,00,000 in five years. WAP to find the population after 10 years.

```python
from sympy import *

r,t,C1 = symbols("r,t,C1")
P = Function("P")(t)

D = Eq(diff(P,t),r*P)

gen = dsolve(D)

display(gen)

c_1 = gen.subs({t:0,P:50000})
c_2 = gen.subs({t:5,P:100000})
C = solve([c_1,c_2],[C1,r],manual = True,dict = True)

p_t = gen.subs(C[-1])
display(p_t)

print("Population after 10 years-")
display(p_t.subs({t:10}))

```

> If a population of a city has doubled from past $25$ years from $1$ lakh, WAP to find when the population will reach $5$ lakh.

```python
from sympy import *

r,t,C1 = symbols("r,t,C1")
P = Function("P")(t)

D = Eq(diff(P,t),r*P)

gen = dsolve(D)

constraints = solve([gen.subs({t:0,P:100000}),gen.subs({t:25,P:200000})],[C1,r],dict = True)[-1]

Pt = gen.subs(constraints)
display(Pt)

print("Time taken is -")
T = Eq(t,float(solve(Pt.subs({P:500000}),t)[0]))
display(T)
```

**Logistic Growth Mode**
> Logistic growth is a type of exponential growth with a limit to the maximum value of the function. WAP to implement a logistic growth model. Solve for function representing quantity over time. The differential equation for such a growth is given as follows $\ldots$<br>
$\quad \quad \dfrac{dN}{dt}\propto rN\left(1 - \dfrac{N}{M}\right)$<br>
where $\ldots$<br>
$\quad \quad N$ is time dependent function giving population as output,<br>
$\quad \quad r$ is the growth rate,<br>
$\quad \quad M$ is the maximum population,<br> 
$\quad \quad t$ is time.

```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

t,r,M,N0,C1 = symbols("t,r,M,N0,C1")

N = Function("N")(t)

D = Eq(diff(N,t),r*N*(1 - (N/M)))
C = solve(dsolve(D).subs({t:0,N:N0}),C1)[0]

N_t = dsolve(D).subs({C1:C}).simplify()
print("The Equation for logistic growth is-")
display(N_t)
```

```python
#constrains = {
#    M: 1000,
#    r: 1,
#    N0: 100
#}
constrains = {
    M: 500,
    r: 2,
    N0: 2
}

Nt = N_t.subs(constrains)

X = np.linspace(0,15,100)
Y = lambda X: [Nt.subs({t: i}).rhs for i in X]

plt.xlabel("TIME")
plt.ylabel("POPULATION")
plt.title("PLOT OF LOGISTIC GROWTH")
plt.axvline(x=0,color = "black")
plt.axhline(y=0,color = "black")
plt.grid()
plt.plot(X,Y(X))
plt.show()
```

> The population of hamsters follows logistic growth. If the population reaches the carrying capacity in $8$ months, WAP to find the rate of growth of hamsters. Given, carrying capacity is $500$ and initially there were $5$ hamsters.

```python
print("Logistic Growth is given as-")
display(N_t)


capacity = 500

constraints = {
    t:8,
    M: capacity,
    N0:5
}

rate = Eq(N_t.subs(constraints).rhs,capacity-1)
print("For given Problem")
display(rate)

print("The rate of growth of hamsters is-")
display(Eq(r,float(solve(rate,r)[-1])))
```
