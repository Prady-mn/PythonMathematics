```python

class Planet():
    #__G = 6.674e-11;
    __G = 0.028;
    dt = 1;
    #__dt = 3e8
    #__M0 = 2e30
    __M0 = 1;
    Position = make_dataclass("Position", [("x", float), ("y", float)])
    Velocity = make_dataclass("Velocity",[("vx",float), ("vy",float)])
    Acceleration = make_dataclass("Acceleration",[("ax",float), ("ay",float)])
    
    def __init__(self,m = 0,position = [0,0],velocity = [0,0],radius = 1):
        """
        Constructor of Class.

        Parameters
        ----------
        m : int
            mass
        position : [x,y] int
            x and y position
        velocity : [vx,vy] float
            x and y velocity
        radius : int
            radius
 
        Returns
        -------
        Object of Planet
        """
        self.m = m*Planet.__M0;
        self.x = position[0];
        self.y = position[1];
        self.vx = velocity[0];
        self.vy = velocity[1];
        self.radius = radius;
        
        
    def displacement(self):
        """
        Displacement from origin
        
        Parameters
        ----------
        self : Planet()
            Object of Planet
        Returns
        ----------
        [x , y]
        """
        return [self.x,self.y];
    
    def distance(self,p = 'Planet'):
        return ((self.x - p.x)**2 + (self.y - p.y)**2)**0.5;
    
    def position(self):
        return [self.x,self.y];
    
    def displacement(self,p = 'Planet'):
        """
        Displacement of self from p
        
        Parameters
        ----------
        p1 : Planet()
            Object of Planet
        
        p1 : Planet()
            Object of Planet
        Returns
        ----------
        [x , y,angle from x - axis ,distance]
        """
        dx = self.x - p.x;
        dy = self.y - p.y;
        if(dx == 0):
            return [dx, dy, copysign(pi/2,dy),((dx)**2 + (dy)**2)**0.5];
        return [dx, dy,np.arctan(dy/dx),((dx)**2 + (dy)**2)**0.5];
    
    def velocity(self):
        return [self.vx,self.vy];
    
    def speed(self):
        return (self.vx**2 + self.vy**2)**0.5
    
    def force(self,p = 'Planet'):
        d = p.displacement(self)
        #print("Distance: ",d)
        F = Planet.__G*(self.m*p.m)/(d[-1]**2)
        
        Fx = F*d[0]/d[-1]
        Fy = F*d[1]/d[-1]
        return [Fx,Fy,F]
    
    def set_dT(t):
        Planet.__dt = t;
        
    def update(p1 = 'Planet',p2 = 'Planet',dt = 1):
        F = p2.force(p1);
        #print("Force:",F)
        #print("Velocity: ",p2.velocity())
        #d = p1.displacement(p2);
        #for i in [p1,p2]:
        p2.x = p2.x + p2.vx*dt + 0.5*(F[0]/p2.m)*dt**2
        p2.y = p2.y + p2.vy*dt + 0.5*(F[1]/p2.m)*dt**2
        p2.vx = p2.vx + (F[0]/p2.m)*dt
        p2.vy = p2.vy + (F[1]/p2.m)*dt
        
    def orbitalVelocity(self,p='Planet',a = 1):
        return ((Planet.__G*(self.m + p.m)*((2/self.distance(p)) - (1/a)))**0.5)
            
```
