```python
import matplotlib.pyplot as plt
import scipy.interpolate as intp
import scipy.optimize as optm
import numpy as np
mu = np.array([1.633,1.624,1.606,1.592,1.589][::-1])
lamda_inv_sq = np.array(sorted([i**-2 for i in [4078,4358,5461,5780,6350]]))
x = np.linspace(lamda_inv_sq.min(),lamda_inv_sq.max(),100)
f = intp.make_interp_spline(lamda_inv_sq,mu)

curve = np.polyfit(lamda_inv_sq,mu,1)
f2 = np.poly1d(curve)

plt.scatter(lamda_inv_sq,mu)
plt.plot(x,f(x))
plt.plot(lamda_inv_sq,f2(lamda_inv_sq))

wavelengths = dict(zip(["Red","Blue-Green","Blue","Violet"],[np.roots(f2 - i)[0]**-0.5 for i in [1.595,1.599,1.611,1.616]]))
print(wavelengths)
plt.grid()
plt.show()
```
